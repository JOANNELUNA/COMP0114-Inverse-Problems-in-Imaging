The size of the sinogram is determined by the number of projection angles and the number of projection positions for each angle (which is the same as the length of the diagonal of the image, rounded up to the nearest integer). In this case, since the Shepp-Logan phantom has a size of $(128, 128)$, the length of the diagonal is $\sqrt{128^2 + 128^2} \approx 181$, so the sinogram has a size of $(181, 180)$. Since the radon function rounds the diagonal length up to the nearest integer, the actual number of projection positions is $182$ as shown above. The number of projection angles is $180$, since we are using 1-degree intervals from 0 to $179$ degrees.

Therefore, the output sinogram has size $(182, 180)$. However, since the "circle = True" is applied in the radon function, the sinogram is cropped to a square shape, resulting in a final size of $(128, 180)$.

### 2. Calculate an explicit matrix form of the Radon transform and investigate its SVD.
You can construct an explicit matrix form of the Radon transform by calling it on an image with a single pixel of value 1 and the rest 0.
To do this :
* Create a zero matrix $A$ with number of rows equal to the number of angles $Ã—$ number of projection samples, and with number of columns equal to the number of pixels in the image.

* Go through a loop for the number of pixels in the image (in column-major). For each pixel $j$ in the loop create an image of the same size with a 1 in that pixel and zero everywhere else. Take the Radon transform of this image and reshape the result into a column vector. This forms the $j^th$ column of the matrix $A$.

* Carry out the above steps for a manageable size of matrix, e.g. a $64 Ã— 64$ image and only 45 projections. Take the SVD of this matrix explicitly and investigate how the SVD spectrum (the singular values) varies with

(1)  Keeping the range of angles from $0 â†’ 180$ but varying the number of projections. E.g angles = $[0:4:179]$ gives 45 projections at $4$ degree seperation.

```python
def SVD_ProjectionNo(image_size, num_angles_list):
    # Define the number of angles and projection samples
    num_samples = image_size
    svd_spectrum = []
    for num_angles in num_angles_list:
        # Create a zero matrix with the appropriate size
        A = np.zeros((num_angles*num_samples, image_size ** 2))
        
        # Loop over each pixel in the image
        for j in range(image_size ** 2):
            # Create an image with a single pixel of value 1 at the jth position
            f = np.zeros((image_size, image_size))
            f[j // image_size, j % image_size] = 1
            
            # Take the Radon transform of the image and reshape it into a column vector
            radon_transform = radon(f, theta=np.linspace(0, 180, num_angles, endpoint=False), circle=True)
            # A[:,j] = np.reshape(radon_transform, (num_angles*num_samples,))
            # A[:, j] = radon_transform.reshape(-1)
            column = radon_transform.reshape(-1, 1)

            # Add the column to the matrix A
            A[:, j] = column.flatten() 

        # Take the SVD of the matrix A
        U, S, Vt = np.linalg.svd(A)

        # Plot the singular values
        plt.plot(S, label='Number of projections = {}'.format(num_angles))
        plt.legend()

        # Add the singular values to the list
        svd_spectrum.append(S)
        
    plt.title('SVD Spectrum vs Number of Projections')
    plt.xlabel('Singular Value Index')
    plt.ylabel('Singular Value')
    plt.show() 
    

    # Plot the singular values against the number of projections
    fig, axs = plt.subplots(4, 3, figsize=(15, 20))
    for i, spectrum in enumerate(svd_spectrum):
        row = i // 3
        col = i % 3
        axs[row, col].plot(spectrum, label=f'Number of projections = {num_angles_list[i]}')
        axs[row, col].legend()
        axs[row, col].set_xlabel('Singular Value Index')
        axs[row, col].set_ylabel('Singular Value')
    plt.show()
```

```python
image_size = 64
num_angles_list = [5, 9, 12, 15, 18, 20, 30, 36, 45, 60, 90, 180]
SVD_ProjectionNo(image_size, num_angles_list)
```

The singular value decomposition (SVD) of the explicit matrix form of the Radon transform provides insights into image's properties and structure. In the case of the Radon transform. When varying the range of angles and number of projections, the amount of information available to reconstruct the image is essentially changed. The number of projections determines the amount of information about the object's geometry, while the range of angles determines the coverage of projections. As the number of projections or the range of angles is increased, more information about the object's structure is captured, resulting in a larger number of singular values in the SVD spectrum. 

The graph above shows the distribution of singular values of the Radon transform matrix for different numbers of projections (i.e. angles) in the range 0 to 180 degrees.

* As we increase the number of projections, the range of singular values increases, which indicates that there are more non-zero singular values. This is expected, as a larger number of projections means that more information is available, resulting in a higher rank and more singular values. The different endings of the lines in the plot reflect these changes in the distribution of singular values as the number of projections changes.

* Additionally, we can see that as the number of projections increases, the decay rate of the singular values becomes less rapid. This means that there are more singular values contributing to the reconstruction of the image and that the reconstruction is becoming more accurate. As more projections are added, the amount of information available about the object's geometry increases. This additional information helps to reduce ambiguity in the reconstruction process, resulting in a more spread-out SVD spectrum. As a result, the spectrum becomes less sensitive to changes in the number of projections, and the frequency at which the singular values decay becomes less rapid.

* Finally, we can observe that the distribution of singular values is not uniform, with a few singular values dominating the others. This indicates that the matrix has a few important features that capture most of the information about the image.

(2) Keeping the number of projection the same, but varying the range to being less than $0 â†’ 180$ (i.e. limited angle). E.g. angles = $[0:1:44]$ gives 45 projections over only a 45 degree range.

```python
def SVD_AngleRange(image_size, max_angles_list):
    # Define the number of projection samples
    num_samples = image_size
    num_angles = 45
    svd_spectrum = []
    
    for max_angles in max_angles_list:        
        # Create a zero matrix with the appropriate size
        A = np.zeros((num_angles*num_samples, image_size ** 2))
        
        # Loop over each pixel in the image
        for j in range(image_size ** 2):
            # Create an image with a single pixel of value 1 at the jth position
            f = np.zeros((image_size, image_size))
            f[j // image_size, j % image_size] = 1
            
            # Take the Radon transform of the image and reshape it into a column vector
            radon_transform = radon(f, theta=np.linspace(0, max_angles, num_angles, endpoint=False), circle=True)
            column = radon_transform.reshape(-1, 1)

            # Add the column to the matrix A
            A[:, j] = column.flatten() 

        # Take the SVD of the matrix A
        U, S, Vt = np.linalg.svd(A)

        # Plot the singular values
        plt.plot(S, label='Maximum angle = {} degrees'.format(max_angles))
        plt.legend()

        # Add the singular values to the list
        svd_spectrum.append(S)
        
    plt.title('SVD Spectrum vs Range of Angles')
    plt.xlabel('Singular Value Index')
    plt.ylabel('Singular Value')
    plt.show()

    # Plot the singular values against the number of projections
    fig, axs = plt.subplots(4, 3, figsize=(15, 20))
    for i, spectrum in enumerate(svd_spectrum):
        row = i // 3
        col = i % 3
        axs[row, col].plot(spectrum, label='Maximum angle = {} degrees'.format(max_angles))
        axs[row, col].legend()
        axs[row, col].set_xlabel('Singular Value Index')
        axs[row, col].set_ylabel('Singular Value')
    plt.show()
```

```python
image_size = 64
max_angles_list = np.linspace(15, 180, 12, endpoint=True)
SVD_AngleRange(image_size, max_angles_list)
```

The graph of the SVD spectrum shows how the singular values of the Radon transform matrix vary with the range of angles used to generate the projection data. 

* It can be observed that all the lines in the plot end at the same point when the range of angles changes. This is in the expectation that changing the range of angles does not affect the rank of the Radon transform matrix. The rank of the matrix is determined by the number of pixels in the image, and this does not change with the range of angles.

* Also, when the number of projections is fixed, increasing the range of angles results in a decrease in the frequency of the singular values which are smaller than 500, and an increase in the frequency of the singular values which are larger than 500. a reduction in ambiguity during the reconstruction process, resulting in more well-distributed singular values across the spectrum. In particular, the smaller singular values, which correspond to high-frequency information in the reconstructed image, tend to decrease in frequency because they are more sensitive to noise and errors in the measurements. On the other hand, the larger singular values, which correspond to low-frequency information in the reconstructed image, tend to increase in frequency because they are less sensitive to noise and errors in the measurements and provide more reliable information about the object's structure.

* Furthermore, he frequency of the singular values dropping to zero at different singular values and at different speeds. This can be attributed to the amount of information captured by the Radon transform. As the range of angles increases, more information about the object's geometry is captured, resulting in a larger spread of singular values. The larger spread leads to a faster drop in frequency of the smaller singular values, as they capture less significant information about the object's geometry. On the other hand, the slower drop in frequency of smaller singular values for the smaller range of angles is due to the limited information captured, resulting in fewer singular values in the spectrum. The drop in frequency occurs at different singular values because the singular values correspond to the amount of information captured by each singular vector. Therefore, the amount and quality of information captured by the Radon transform affects the distribution and frequency of singular values, which in turn affects the reconstruction process.

### 3. Implement a matrix-free regularised least-squares solver for the Radon Transform.
You can find a regularised solution to the inverse Radon Transform by solving
$$(A^TA + Î±L)f_{âˆ—} = A^Tg$$
where $A$ is the Radon transform operator, $g$ is the measured sinogram, $f_*$ is the image to be reconstructed, $L$ is the regularization matrix, and $\alpha$ is the regularization parameter. Implement the matrix-free Krylov solver will be used to solve this question.

```python
from scipy.sparse.linalg import minres
from scipy import ndimage

def matrix_free_solver(sinogram, theta, alpha, order=0, x0=None, maxiter=100):
    """
    Solve the regularized least-squares problem using a matrix-free solver.
    
    sinogram: ndarray
        The sinogram of the image.
    theta: ndarray
        The projection angles in degrees.
    alpha: float
        The regularization parameter.
    order: int
        The order of Tikhonov regularization. Can be 0 or 1.
    x0: ndarray or None
        The initial guess for the image. If None, a zero array will be used.
    maxiter: int
        The maximum number of iterations.
    
    Returns:
    -------
    x: ndarray
        The reconstructed image.
    """
    n = sinogram.shape[0]
    # m = int(np.sqrt(2) * n)
    
    # Define the forward and adjoint operators
    def A(x):
        if x.ndim == 2:
            x_2d = x
        elif x.ndim == 1:
            x_2d = x.reshape((128,int(x.size/128)))
        return radon(x_2d, theta=theta, circle=True)#.ravel()

    def AT(x):
        if x.ndim == 2:
            x_2d = x
        elif x.ndim == 1:
            x_2d = x.reshape((128,int(x.size/128)))
        return iradon(x_2d, theta=theta, circle=True, filter_name='ramp')#.ravel()

    
    # Define the regularization operator
    if order == 0:
        def L(x):
            return alpha*np.sqrt(n)*x.ravel()
    elif order == 1:
        def L(x):
            return alpha*ndimage.laplace(x).ravel()
            # laplacian = np.zeros_like(x)
            # laplacian[1:-1, 1:-1] = x[:-2, 1:-1] - 2*x[1:-1, 1:-1] + x[2:, 1:-1] + \
            #                        x[1:-1, :-2] - 2*x[1:-1, 1:-1] + x[1:-1, 2:]
            # return alpha*laplacian.ravel()
    else:
        raise ValueError('Invalid value for order parameter.')
    
    # Define the Ax and b functions for Ax = b
    def Ax(x):
        "A - LHS of function"
        return AT(A(x)) + L(x)
    
    def b(y):
        "b - RHS of function"
        return AT(y)
    
    # Initialize the solution vector
    if x0 is None:
        x0 = np.zeros((n, n))
    x = x0.ravel()
    
    # Use the conjugate gradient method to solve the system of linear equations
    Ax_op = LinearOperator((n,n), matvec=Ax)
    print(Ax_op)
    print(b(sinogram).shape)
    print(sinogram.shape)
    x, info = cg(Ax_op, b(sinogram), x0=np.zeros((1, 1)), maxiter=maxiter, tol=1e-6, atol=0)
    # x, info = minres(Ax, b(sinogram), x0=x, maxiter=maxiter, tol=1e-6, shift=0, show=False)

    # Reshape the solution vector to an image
    x_2d = x.reshape((n, n))
    
    return x_2d
```

(3) Write a function that implements thresholding for a given range (the different scales of your wavelet coefficients) and threshold parameter, and form a modified image by performing the inverse wavelet transform on the thresholded coefficients.


```python
def threshold(wave_coeffs, threshold_val, threshold_range):
    """
    Apply thresholding to the wavelet coefficients.

    Args:
    wave_coeffs: 2D numpy array, containing wavelet coefficients
    threshold_val: float, the threshold value for hard thresholding
    threshold_range: list of integers, indicating the scales to threshold

    Returns:
    2D numpy array, modified wavelet coefficients
    """
    # Copy the input wavelet coefficients
    modified_coeffs = wave_coeffs.copy()

    # Loop through each scale in threshold range
    for scale in threshold_range:
        # Apply hard thresholding to the detail coefficients
        modified_coeffs[1][scale][abs(modified_coeffs[1][scale]) < threshold_val] = 0
    
    return modified_coeffs
```

(4) Create a noisy version of your original image and perform denoising by thresholding of the wavelet coefficients. Investigate the effect of changing the range and the threshold parameter.

```python
def denoise_wavelet(image, threshold_val, threshold_range):
    """
    Denoise image using wavelet thresholding.

    Args:
    image: 2D numpy array, containing the original image
    threshold_val: float, the threshold value for hard thresholding
    threshold_range: list of integers, indicating the scales to threshold

    Returns:
    2D numpy array, denoised image
    """
    # Calculate the wavelet transform of the image
    wave_coeffs = pywt.wavedec2(image, 'haar', level=3)

    # Apply thresholding to the wavelet coefficients
    modified_coeffs = threshold(wave_coeffs, threshold_val, threshold_range)

    # Reconstruct the image using the modified wavelet coefficients
    denoised_image = pywt.waverec2(modified_coeffs, 'haar')

    return denoised_image
```
